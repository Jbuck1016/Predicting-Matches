```python
import pandas as pd

# URLs of the CSV files
urls = [
    "https://raw.githubusercontent.com/Jbuck1016/Predicting-Matches/main/Bundesliga_fbref_poss_stats.csv",
    "https://raw.githubusercontent.com/Jbuck1016/Predicting-Matches/main/EPL_fbref_poss_stats.csv",
    "https://raw.githubusercontent.com/Jbuck1016/Predicting-Matches/main/Ligue_Une_fbref_poss_stats.csv",
    "https://raw.githubusercontent.com/Jbuck1016/Predicting-Matches/main/Serie_A_fbref_poss_stats.csv",
    "https://raw.githubusercontent.com/Jbuck1016/Predicting-Matches/main/la_liga_fbref_poss_stats.csv"
]

# Load each CSV file into a DataFrame and store them in a list
dataframes = [pd.read_csv(url) for url in urls]

# Combine all DataFrames into a single DataFrame
combined_df = pd.concat(dataframes, ignore_index=True)
# Drop rows where the 'round' column contains the word 'round'
combined_df = combined_df[~combined_df['round'].str.contains('round', case=False, na=False)]


# Display the combined DataFrame
print(combined_df.head())  # Shows the first few rows of the combined DataFrame
```

```python
import pandas as pd

# Assuming 'combined_df' is already your loaded DataFrame with the necessary columns

# Define a function that assigns home and away teams based on the venue
def assign_teams(row):
    if row['venue'] == 'Home':
        home_team = row['team']
        away_team = row['opponent']
    else:
        home_team = row['opponent']
        away_team = row['team']
    return pd.Series([home_team, away_team])

# Apply the function along the rows (axis=1)
combined_df[['home_team', 'away_team']] = combined_df.apply(assign_teams, axis=1)

# Print the DataFrame to verify the new columns
print(combined_df[['league', 'season', 'date', 'home_team', 'away_team', 'venue', 'result', 'GF', 'GA']])
```

```python
import pandas as pd

# URL of the CSV file
url = 'https://raw.githubusercontent.com/Jbuck1016/Predicting-Matches/main/big_5_shot_stats.csv'

# Load the CSV file into a DataFrame
shots_df = pd.read_csv(url)


# Select only the necessary columns from shots_df
shots_df = shots_df[['date', 'home_team', 'away_team', 'Shots', 'Expected Goals Minus Goals']]

# Merge the DataFrames on the date, home_team, and away_team
combined_df = combined_df.merge(shots_df, on=['date', 'home_team', 'away_team'], how='left')

# Display the first few rows to confirm the merge
print(combined_df.head())
```

```python
import pandas as pd

# Define the function to enhance the game stats in the DataFrame
def enhance_game_stats(combined_df):
    # Initialize new columns
    combined_df['Home_Touches_Att 3rd'] = None
    combined_df['Home_Touches_Att Pen'] = None
    combined_df['Away_Touches_Att 3rd'] = None
    combined_df['Away_Touches_Att Pen'] = None
    
    # Initialize columns for Shots and Expected Goals Minus Goals
    combined_df['Home_Shots'] = None
    combined_df['Away_Shots'] = None
    combined_df['Home_Expected Goals Minus Goals'] = None
    combined_df['Away_Expected Goals Minus Goals'] = None

    # For each unique game, populate the home and away stats across both rows
    for game in combined_df['game'].unique():
        # Subset DataFrame for the specific game
        game_df = combined_df[combined_df['game'] == game]
        
        # Assuming there are always exactly two rows per game
        for index, row in game_df.iterrows():
            if row['venue'] == 'Home':
                home_stats = row
                home_index = index
            else:
                away_stats = row
                away_index = index
        
        # Apply home and away stats to both rows
        combined_df.loc[game_df.index, 'Home_Touches_Att 3rd'] = home_stats['Touches_Att 3rd']
        combined_df.loc[game_df.index, 'Home_Touches_Att Pen'] = home_stats['Touches_Att Pen']
        combined_df.loc[game_df.index, 'Away_Touches_Att 3rd'] = away_stats['Touches_Att 3rd']
        combined_df.loc[game_df.index, 'Away_Touches_Att Pen'] = away_stats['Touches_Att Pen']

        # Apply Shots and Expected Goals Minus Goals
        combined_df.loc[game_df.index, 'Home_Shots'] = home_stats['Shots']
        combined_df.loc[game_df.index, 'Away_Shots'] = away_stats['Shots']
        combined_df.loc[game_df.index, 'Home_Expected Goals Minus Goals'] = home_stats['Expected Goals Minus Goals']
        combined_df.loc[game_df.index, 'Away_Expected Goals Minus Goals'] = away_stats['Expected Goals Minus Goals']

    return combined_df

# Assuming combined_df is already defined and contains your data
combined_df = enhance_game_stats(combined_df)

# You can now use enhanced_combined_df for further analysis or export
print(combined_df)
```
