```python
import pandas as pd
import soccerdata as sd

# Initialize the MatchHistory object
mh = sd.MatchHistory(leagues=["ITA-Serie A", "ENG-Premier League", "ESP-La Liga", "FRA-Ligue 1", "GER-Bundesliga"],
                     seasons=["23", "22", "21", "20", "19", "18"])

hist = mh.read_games()
hist.reset_index(inplace=True)
hist.head()
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame loaded with the data

# List of columns you want to keep
columns_to_keep = ['league', 'season', 'game', 'date', 'home_team', 'away_team', 'B365H', 'B365D', 'B365A']

# Drop all other columns
hist = hist[columns_to_keep]
import pandas as pd

# Assuming 'df' is your DataFrame

# Convert 'date' to datetime if it's not already
hist['date'] = pd.to_datetime(hist['date'])

# Extract only the date part
hist['date'] = hist['date'].dt.date


# Now 'df' only contains the columns specified
hist
```
