```python
import soccerdata as sd
import pandas as pd
fbref = sd.FBref(leagues=["ITA-Serie A", "ENG-Premier League", "ESP-La Liga", "FRA-Ligue 1", "GER-Bundesliga"], seasons=["23", "22", "21", "20", "19", "18", "17"])
team_match_stats_22 = fbref.read_team_match_stats(stat_type="shooting")
team_match_stats_22.head()
```

```python
# Flattening the MultiIndex columns
team_match_stats_22.columns = ['_'.join(col).strip('_') if col[1] else col[0] for col in team_match_stats_22.columns]

# Print the DataFrame columns to verify the changes
print(team_match_stats_22.columns)
```

```python
team_match_stats_22
```

```python
# Resetting the index to turn the MultiIndex into columns
team_match_stats_22 = team_match_stats_22.reset_index()

# Step 1: Rename the column
team_match_stats_22.rename(columns={'Expected_G-xG': 'Expected Goals Minus Goals', 'Standard_Sh': 'Shots'}, inplace=True)

# Step 2: Create 'home_team' and 'away_team' columns
team_match_stats_22['home_team'] = team_match_stats_22.apply(lambda row: row['team'] if row['venue'] == 'Home' else row['opponent'], axis=1)
team_match_stats_22['away_team'] = team_match_stats_22.apply(lambda row: row['opponent'] if row['venue'] == 'Home' else row['team'], axis=1)

# Step 3: Filter and drop unnecessary columns
final_columns = ['date', 'team', 'home_team', 'away_team', 'opponent', 'Shots', 'Expected Goals Minus Goals']
team_match_stats_22 = team_match_stats_22[final_columns]

# Display the resulting DataFrame
print(team_match_stats_22)
```

```python
team_match_stats_22.to_csv('big_5_shot_stats.csv', index=False)
```
