```python
import soccerdata as sd
import pandas as pd
fbref = sd.FBref(leagues=["ENG-Premier League"], seasons=["23", "22"])
team_match_stats_22 = fbref.read_team_match_stats(stat_type="possession")

fbref = sd.FBref(leagues=["ENG-Premier League"], seasons=["21", "20"])
team_match_stats_20 = fbref.read_team_match_stats(stat_type="possession")

fbref = sd.FBref(leagues=["ENG-Premier League"], seasons=["19", "18"])
team_match_stats_18 = fbref.read_team_match_stats(stat_type="possession")

fbref = sd.FBref(leagues=["ENG-Premier League"], seasons=["17"])
team_match_stats_17 = fbref.read_team_match_stats(stat_type="possession")
team_match_stats_17.head()
```

```python
combined_df = pd.concat([team_match_stats_17, team_match_stats_18, team_match_stats_20, team_match_stats_22])

# Reset the index if necessary (in case of duplicate indices)
combined_df.reset_index(drop=True, inplace=True)

# Print the combined DataFrame to verify
print(combined_df)
```

```python
# Flattening the MultiIndex columns
combined_df.columns = ['_'.join(col).strip('_') if col[1] else col[0] for col in combined_df.columns]

# Print the DataFrame columns to verify the changes
print(combined_df.columns)
```

```python
combined_df.to_csv('EPL_fbref_poss_stats.csv', index=False)
```
